```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def generalized_euclidean_distance(u, v, w=np.array([1])):
    result = sum(
        abs(w) * ((w >= 0) * abs(u - v) + (w < 0) * (-abs(u - v) + 2)) ** 2
    ) ** (1 / 2)
    return result


w = np.array([-1, 1])
u = np.array([1, 0])
v = np.array([-1, 0])

# u = np.array([-1.5, 0, -1])
# v = np.array([0, 0, 1])
# x = np.array([1, 0, 1])

print(generalized_euclidean_distance(u, v, w))
```

```python
u_values = np.arange(-1, 1.1, 0.1)
v_values = np.arange(-1, 1.1, 0.1)

results = np.zeros((len(u_values), len(v_values)))

for i, u in enumerate(u_values):
    for j, v in enumerate(v_values):
        result = generalized_euclidean_distance(
            np.array(u), np.array(v), w=np.array([1])
        )
        results[i, j] = result

plt.figure(figsize=(8, 6))
plt.imshow(
    results, cmap="viridis", extent=[-1, 1, -1, 1], origin="lower", aspect="auto"
)
plt.colorbar(label="Distance")
plt.xlabel("u")
plt.ylabel("v")
plt.title("Generalized Euclidean Distance Heatmap")
plt.show()
```
