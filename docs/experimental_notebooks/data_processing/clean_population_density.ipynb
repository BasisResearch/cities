```python
counties_area = pd.read_csv("../../../../cities/data/raw/LND01.csv")
population = pd.read_csv("../../../../cities/data/processed/population_wide.csv")

# 5-digit FIPS code
counties_area['GeoFIPS'] = counties_area['STCOU'].astype(str).str.zfill(5)
population['GeoFIPS'] = population['GeoFIPS'].astype(str).str.zfill(5)

counties_area['area_sqm'] = counties_area['LND010190D']

population_density = population.copy()

# compute population density 
for index, row in population.iterrows():
    geofips = row['GeoFIPS']
    area_row = counties_area[counties_area['GeoFIPS'] == geofips]['area_sqm']
    if area_row.size == 0:
        print('No area found for {}'.format(geofips))
    else:
        # iterate over all year columns
        for col in population.columns[2:]:
            tmp = row[col] / area_row.astype(float)
            population_density.loc[index, col] = tmp.iloc[0]


# population_density.to_csv("../../../../cities/data/processed/population_density_wide.csv")
population_density
```
